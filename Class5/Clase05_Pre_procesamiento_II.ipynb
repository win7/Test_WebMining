# 1. Identificación de Lenguaje
Librería langid: https://github.com/saffsd/langid.py

```python
!pip install langid
```

```python
import langid
```

```python
langid.classify('How are you?')
```

```python
from langid.langid import LanguageIdentifier, model
id_model = LanguageIdentifier.from_modelstring(model, norm_probs=True)
```

```python
print(id_model.classify('Hola ¿cómo estás?'))
print(id_model.classify('Hola ¿cómo estás? te escribia para preguntarte sobre'))
print(id_model.classify('Hola ¿cómo estás? te escribia para preguntarte sobre la clase del sábado'))
print(id_model.classify('How are you?'))
```

```python
id_model.classify('sala de estar'), id_model.classify('livingroom')
```

```python
print(id_model.classify('Tengo ganas de comer un sandwich'))
print(id_model.classify('comer un sandwich'))
```
